## Data Collection

### 1. Extracting the population of each state:

```python
# Connecting to webdriver
driver=webdriver.Chrome(r"C:\Users\Lenovo\chromedriver_112.exe")
time.sleep(1)

# Opening 'https://en.wikipedia.org/wiki/List_of_states_in_India_by_past_population' in chrome browser:
url='https://en.wikipedia.org/wiki/List_of_states_in_India_by_past_population'
driver.get(url)
driver.maximize_window()
time.sleep(2)
```

```python
# Creating Empty List:
POP_01 =[]
POP_11 =[]
State_UT =[]

            # Extracting 'Population 2001 Census' into the Empty list: 
try:
    pop_01=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[8]')
    for i in pop_01:
        POP_01.append(i.text)
except NoSuchElementException:
    POP_01.append('NA')
except StaleElementReferenceException:
    POP_01.append('NA')    
# POP_01 = POP_01[0:-1]

            # Extracting 'Population 2011 Census'into the Empty list:
try:
    pop_11=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[9]')
    for i in pop_11:
        POP_11.append(i.text)
except NoSuchElementException:
    POP_11.append('NA')
except StaleElementReferenceException:
    POP_11.append('NA')
# POP_11 = POP_11[0:-1]

            # Extracting 'State_UT'into the Empty list:
try:
    state_UT=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[2]/a')
    for i in state_UT:
        State_UT.append(i.text)
except NoSuchElementException:
    State_UT.append('NA')
except StaleElementReferenceException:
    State_UT.append('NA')   
# State_UT = State_UT[0:-1]
```

```python
len(POP_01),len(POP_11), len(State_UT)
```

```python
# Dataframe for scrap data:

Wiki_POP = pd.DataFrame({'State/UT':State_UT, 'POP_2001':POP_01, 'POP_2011':POP_11})
print('\033[1m'+'Census of India(2001 & 2011) from Wikipedia :'+'\033[0m')
#Wiki_POP = Wiki_POP.set_index('State/UT')
Wiki_POP
```

```python
# saving the dataframe
Wiki_POP.to_csv('Census of India_2001_2011 from Wikipedia.csv')
```

### 2. Extracting the Literacy Rate in each state:

```python
# Connecting to webdriver
driver=webdriver.Chrome(r"C:\Users\Lenovo\chromedriver_112.exe")
time.sleep(1)

# Opening 'https://www.rbi.org.in/Scripts/PublicationsView.aspx?id=19995' in chrome browser:
url='https://www.rbi.org.in/Scripts/PublicationsView.aspx?id=19995'
driver.get(url)
driver.maximize_window()
time.sleep(2)
```

```python
# Creating Empty List:
LR_01 =[]
LR_11 =[]
State_UT =[]

            # Extracting 'Population 2001 Census' into the Empty list: 
try:
    lr_01=driver.find_elements(By.XPATH,'/html/body/form/div[5]/div[3]/div/table[2]/tbody/tr[4]/td/table/tbody/tr/td[7]')
    for i in lr_01:
        LR_01.append(i.text)
except NoSuchElementException:
    LR_01.append('NA')
except StaleElementReferenceException:
    LR_01.append('NA')    
LR_01 = LR_01[1:-1]

            # Extracting 'Population 2011 Census'into the Empty list:
try:
    lr_11=driver.find_elements(By.XPATH,'/html/body/form/div[5]/div[3]/div/table[2]/tbody/tr[4]/td/table/tbody/tr/td[8]')
    for i in lr_11:
        LR_11.append(i.text)
except NoSuchElementException:
    LR_11.append('NA')
except StaleElementReferenceException:
    LR_11.append('NA')
LR_11 = LR_11[1:-1]

            # Extracting 'State_UT'into the Empty list:
try: 
    state_UT=driver.find_elements(By.XPATH,'/html/body/form/div[5]/div[3]/div/table[2]/tbody/tr[4]/td/table/tbody/tr/td[1]')
    for i in state_UT:
        State_UT.append(i.text)
except NoSuchElementException:
    State_UT.append('NA')
except StaleElementReferenceException:
    State_UT.append('NA')   
State_UT = State_UT[2:-2]
```

```python
len(LR_01),len(LR_11), len(State_UT)
```

```python
# Dataframe for scrap data:

RBI_LR = pd.DataFrame({'State/UT':State_UT, 'LR_2001':LR_01, 'LR_2011':LR_11})
print('\033[1m'+'State-wise Literacy Rate of India(2001 & 2011) from RBI :'+'\033[0m')

RBI_LR
```

```python
# saving the dataframe
RBI_LR.to_csv('State-wise Literacy Rate of India_2001_2011 from RBI.csv')
```

### 3. Extracting the Area of each state in India:

```python
# Connecting to webdriver
driver=webdriver.Chrome(r"C:\Users\Lenovo\chromedriver_112.exe")
time.sleep(1)

# Opening 'https://www.mapsofindia.com/census2011/population.html' in chrome browser:
url='https://www.mapsofindia.com/census2011/population.html'
driver.get(url)
driver.maximize_window()
time.sleep(2)
```

```python
# Creating Empty List:
State = []
Area_State = []


            # Extracting 'Population 2001 Census' into the Empty list: 
try:
    state = driver.find_elements(By.XPATH,'/html/body/div[8]/div[1]/div[2]/div[2]/div[1]/div/div[1]/div[4]/table/tbody/tr/td[2]')
    for i in state:
        State.append(i.text)
except NoSuchElementException:
    State.append('NA')
except StaleElementReferenceException:
    State.append('NA')    
State = State[1:]

            # Extracting 'Population 2011 Census'into the Empty list:
try:
    Area = driver.find_elements(By.XPATH,'/html/body/div[8]/div[1]/div[2]/div[2]/div[1]/div/div[1]/div[4]/table/tbody/tr/td[4]')
    for i in Area:
        Area_State.append(i.text)
except NoSuchElementException:
    Area_State.append('NA')
except StaleElementReferenceException:
    Area_State.append('NA')
Area_State = Area_State[1:]           
```

```python
len(Area_State), len(State)
```

```python
# Dataframe for scrap data:

MOI_Area_State = pd.DataFrame({'State/UT':State, 'Area (sq. km)':Area_State})
print('\033[1m'+'Areas of the states in India from Maps Of India:'+'\033[0m')

MOI_Area_State
```

```python
# Creating Empty List:
UT = []
Area_UT =[]


 # Extracting 'Population 2001 Census' into the Empty list: 
try:
    ut = driver.find_elements(By.XPATH,'/html/body/div[8]/div[1]/div[2]/div[2]/div[1]/div/div[1]/table[6]/tbody/tr/td[2]')
    for i in ut:
        UT.append(i.text)
except NoSuchElementException:
    UT.append('NA')
except StaleElementReferenceException:
    UT.append('NA')    
UT = UT[1:]

            # Extracting 'Population 2011 Census'into the Empty list:
try:
    area = driver.find_elements(By.XPATH,'/html/body/div[8]/div[1]/div[2]/div[2]/div[1]/div/div[1]/table[6]/tbody/tr/td[3]')
    for i in area:
        Area_UT.append(i.text)
except NoSuchElementException:
    Area_UT.append('NA')
except StaleElementReferenceException:
    Area_UT.append('NA')
Area_UT = Area_UT[1:]
```

```python
len(Area_UT), len(UT)
```

```python
# Dataframe for scrap data:

MOI_Area_UT = pd.DataFrame({'State/UT':UT, 'Area (sq. km)':Area_UT})
print('\033[1m'+'Areas of the Union Territories in India from Maps Of India:'+'\033[0m')

MOI_Area_UT
```

```python
# Combining two dataframes:
frames = [MOI_Area_State, MOI_Area_UT]
  
MOI_Area = pd.concat(frames)
MOI_Area.reset_index(inplace = True, drop = True)
display(MOI_Area)
```

```python
# saving the dataframe
MOI_Area.to_csv('Areas of the States_UT in India from Maps Of India.csv')
```
